```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import torch
from transformers import (
    PegasusForConditionalGeneration,
    PegasusTokenizer,
    Trainer,
    TrainingArguments,
    DataCollatorForSeq2Seq
)
from datasets import Dataset
from rouge_score import rouge_scorer
import nltk
import re
from tqdm import tqdm
```

```python
nltk.download('punkt')
```

```python
class NewsSummarizationDataset:
    def __init__(self, tokenizer, max_source_length=1024, max_target_length=128):
        self.tokenizer = tokenizer
        self.max_source_length = max_source_length
        self.max_target_length = max_target_length

    def preprocess_text(self, text):
        # Basic text cleaning
        text = re.sub(r'\s+', ' ', text)  # Remove extra whitespace
        text = re.sub(r'[^\w\s.,!?-]', '', text)  # Remove special characters
        return text.strip()

    def prepare_data(self, df):
        # Clean column names by stripping whitespace
        df.columns = df.columns.str.strip()

        # Print column names for debugging
        print("Available columns:", df.columns.tolist())

        # Map column names (handle potential variations)
        content_col = [col for col in df.columns if 'Content' in col][0]
        summary_col = [col for col in df.columns if 'Human Summary' in col][0]

        # Clean the text data
        df['cleaned_content'] = df[content_col].apply(self.preprocess_text)
        df['cleaned_summary'] = df[summary_col].apply(self.preprocess_text)

        # Convert to dataset format
        dataset_dict = {
            'content': df['cleaned_content'].tolist(),
            'summary': df['cleaned_summary'].tolist()
        }
        return Dataset.from_dict(dataset_dict)

    def tokenize_function(self, examples):
        # Tokenize inputs
        model_inputs = self.tokenizer(
            examples['content'],
            max_length=self.max_source_length,
            padding='max_length',
            truncation=True
        )

        # Tokenize targets
        labels = self.tokenizer(
            examples['summary'],
            max_length=self.max_target_length,
            padding='max_length',
            truncation=True
        )

        model_inputs['labels'] = labels['input_ids']
        return model_inputs


```

```python
def load_and_split_data(file_path):
    # Read the dataset
    df = pd.read_excel(file_path)
    # Set the row with "Sr. No" (index 0) as the column names
    df.columns = df.iloc[0]

    # Drop the first row (since it's now the column names) and the title row
    df = df.drop([0]).reset_index(drop=True)

    # If you want to also remove the original unnamed header row
    df = df.rename_axis(None, axis=1)
    # Clean column names
    df.columns = df.columns.str.strip()

    # Print column names for verification
    print("Loaded columns:", df.columns.tolist())

    # Split into train, validation, and test sets
    train_df, temp_df = train_test_split(df, test_size=0.4, random_state=42)
    val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

    return train_df, val_df, test_df
```

```python
def evaluate_summaries(references, predictions):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = {
        'rouge1': [], 'rouge2': [], 'rougeL': []
    }

    for ref, pred in zip(references, predictions):
        score = scorer.score(ref, pred)
        for key in scores:
            scores[key].append(score[key].fmeasure)

    # Calculate average scores
    avg_scores = {key: np.mean(values) for key, values in scores.items()}
    return avg_scores
```

```python
def main():
    # Check if CUDA is available
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Initialize tokenizer and model
    model_name = "google/pegasus-xsum"
    tokenizer = PegasusTokenizer.from_pretrained(model_name)
    model = PegasusForConditionalGeneration.from_pretrained(
        model_name,
        gradient_checkpointing=True,  # Enable gradient checkpointing
        use_cache=False  # Disable cache during training
    )
    model = model.to(device)

    # Load and split data
    train_df, val_df, test_df = load_and_split_data("/content/Synthetic News Dataset & Model Details.xlsx")

    # Initialize dataset processor with smaller max lengths
    dataset_processor = NewsSummarizationDataset(
        tokenizer,
        max_source_length=256,  # Further reduced from 512
        max_target_length=32    # Further reduced from 64
    )

    # Prepare datasets with even smaller samples for memory constraints
    train_df = train_df.sample(n=min(len(train_df), 500))   # Reduced from 1000
    val_df = val_df.sample(n=min(len(val_df), 100))         # Reduced from 200
    test_df = test_df.sample(n=min(len(test_df), 100))      # Reduced from 200

    train_dataset = dataset_processor.prepare_data(train_df)
    val_dataset = dataset_processor.prepare_data(val_df)
    test_dataset = dataset_processor.prepare_data(test_df)

    # Tokenize datasets with smaller batch size
    train_dataset = train_dataset.map(
        dataset_processor.tokenize_function,
        batched=True,
        remove_columns=train_dataset.column_names,
        batch_size=16  # Reduced from 32
    )
    val_dataset = val_dataset.map(
        dataset_processor.tokenize_function,
        batched=True,
        remove_columns=val_dataset.column_names,
        batch_size=16  # Reduced from 32
    )

    # Set up training arguments with memory optimization
    training_args = TrainingArguments(
        output_dir="./pegasus-news-summarizer",
        num_train_epochs=1,
        per_device_train_batch_size=2,     # Reduced from 8
        per_device_eval_batch_size=2,      # Reduced from 8
        gradient_accumulation_steps=8,     # Increased from 4
        warmup_steps=50,                  # Reduced from 100
        weight_decay=0.01,
        logging_dir="./logs",
        logging_steps=25,
        eval_strategy="steps",           # Updated from evaluation_strategy
        eval_steps=100,                  # Reduced from 200
        save_steps=100,                  # Reduced from 200
        load_best_model_at_end=True,
        save_total_limit=1,             # Reduced from 2
        fp16=True if torch.cuda.is_available() else False,
        gradient_checkpointing=True,    # Enable gradient checkpointing
        optim="adamw_torch",           # Use memory-efficient optimizer
        max_grad_norm=0.5,             # Add gradient clipping
    )

    # Initialize data collator with smaller padding size
    data_collator = DataCollatorForSeq2Seq(
        tokenizer,
        model=model,
        padding=True,
        max_length=256  # Limit padding length
    )

    # Initialize trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        data_collator=data_collator,
    )

    # Clear CUDA cache before training
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    # Train the model
    trainer.train()

    # Generation function with memory optimization
    def generate_summary(text):
        # Clear cache before generation
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

        inputs = tokenizer(text, max_length=256, truncation=True, return_tensors="pt")
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            summary_ids = model.generate(
                inputs["input_ids"],
                max_length=32,           # Reduced from 64
                min_length=10,           # Reduced from 20
                num_beams=2,
                length_penalty=1.0,
                early_stopping=True
            )
        return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Generate summaries in smaller batches
    test_predictions = []
    test_references = test_df['Human Summary'].tolist()

    # Process test set in smaller batches
    batch_size = 5
    for i in tqdm(range(0, len(test_df), batch_size), desc="Generating summaries"):
        batch = test_df['Content'].iloc[i:i+batch_size].tolist()
        for content in batch:
            try:
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()
                summary = generate_summary(content)
                test_predictions.append(summary)
            except Exception as e:
                print(f"Error generating summary: {e}")
                test_predictions.append("")

    # Calculate ROUGE scores
    rouge_scores = evaluate_summaries(test_references, test_predictions)
    print("\nROUGE Scores:")
    for metric, score in rouge_scores.items():
        print(f"{metric}: {score:.4f}")

if __name__ == "__main__":
    main()

```

### **Testing the model on unseen articles and Compare the results with human-written summaries.**

```python
import torch
from transformers import PegasusTokenizer, PegasusForConditionalGeneration
import numpy as np
import matplotlib.pyplot as plt

# Same test articles as before
test_articles = [
    """SpaceX successfully launched its latest batch of Starlink satellites into orbit from Cape Canaveral Space Force Station in Florida. The Falcon 9 rocket carried 60 satellites, bringing the total number of Starlink satellites in orbit to over 1,500. The mission marked another milestone in SpaceX's goal to provide global internet coverage. The rocket's first stage booster landed successfully on a drone ship in the Atlantic Ocean, marking its eighth successful recovery.""",

    """Scientists have discovered a new species of deep-sea coral in the Pacific Ocean. The fluorescent coral, found at depths of over 2,000 meters, displays unique bioluminescent properties never before seen in marine life. Researchers believe this discovery could lead to breakthroughs in understanding deep-sea ecosystems and potentially aid in developing new biomedical applications.""",

    """A breakthrough in renewable energy technology has been announced by researchers at MIT. The team has developed a new type of solar cell that can achieve up to 40% efficiency in converting sunlight to electricity, significantly higher than current commercial solar panels. The technology uses a novel material structure that captures a broader spectrum of light.""",

    """The World Health Organization has reported significant progress in global vaccination efforts. Recent data shows that childhood vaccination rates have improved by 15% in developing nations over the past year. The success is attributed to improved distribution networks and community outreach programs. However, challenges remain in reaching remote areas.""",

    """A major archaeological discovery has been made in Egypt's Valley of the Kings. Archaeologists have unearthed a previously unknown tomb containing well-preserved artifacts and hieroglyphics dating back to the New Kingdom period. The tomb appears to belong to a high-ranking official and contains numerous artifacts, including jewelry, pottery, and religious items."""
]

# First, load the saved model and tokenizer
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load tokenizer and model from the saved directory
model_name = "google/pegasus-xsum"  # Base model name
saved_model_path = "/content/pegasus-news-summarizer/checkpoint-4"  # Your saved model directory

print("Loading tokenizer...")
tokenizer = PegasusTokenizer.from_pretrained(model_name)

print("Loading model from saved directory...")
model = PegasusForConditionalGeneration.from_pretrained(saved_model_path)
model = model.to(device)
model.eval()

def analyze_summary(original_text, model_summary):
    orig_length = len(original_text.split())
    summary_length = len(model_summary.split())
    compression_ratio = summary_length / orig_length

    sentences = model_summary.split('.')
    avg_sentence_length = np.mean([len(s.split()) for s in sentences if s.strip()])

    return {
        'original_length': orig_length,
        'summary_length': summary_length,
        'compression_ratio': compression_ratio,
        'avg_sentence_length': avg_sentence_length
    }

def test_saved_model():
    results = []
    print("\nGenerating and analyzing summaries...")

    for idx, article in enumerate(test_articles, 1):
        print(f"\nAnalyzing Article {idx}...")

        # Clear CUDA cache if available
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

        # Generate summary
        inputs = tokenizer(article, max_length=256, truncation=True, return_tensors="pt")
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            summary_ids = model.generate(
                inputs["input_ids"],
                max_length=32,
                min_length=10,
                num_beams=2,
                length_penalty=1.0,
                early_stopping=True
            )
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

        # Analyze summary
        analysis = analyze_summary(article, summary)

        results.append({
            'article_id': idx,
            'original_text': article.strip(),
            'model_summary': summary,
            **analysis
        })

        # Print results for each article
        print(f"\nOriginal Text ({analysis['original_length']} words):")
        print(article.strip())
        print(f"\nGenerated Summary ({analysis['summary_length']} words):")
        print(summary)
        print(f"Compression Ratio: {analysis['compression_ratio']:.2f}")
        print(f"Average Sentence Length: {analysis['avg_sentence_length']:.1f} words")
        print("-" * 80)

    # Calculate overall statistics
    avg_compression = np.mean([r['compression_ratio'] for r in results])
    avg_summary_length = np.mean([r['summary_length'] for r in results])

    print("\nOverall Statistics:")
    print(f"Average Compression Ratio: {avg_compression:.2f}")
    print(f"Average Summary Length: {avg_summary_length:.1f} words")

    # Plot summary characteristics
    plt.figure(figsize=(10, 6))
    plt.bar(
        range(1, len(results) + 1),
        [r['compression_ratio'] for r in results],
        alpha=0.6
    )
    plt.axhline(y=avg_compression, color='r', linestyle='--', label='Avg Compression Ratio')
    plt.xlabel('Article')
    plt.ylabel('Compression Ratio')
    plt.title('Summary Compression Ratio by Article')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    return results

# Run the test
results = test_saved_model()
```
