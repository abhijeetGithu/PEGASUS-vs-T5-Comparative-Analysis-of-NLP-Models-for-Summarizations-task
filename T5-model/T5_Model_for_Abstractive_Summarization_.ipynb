```python
def preprocess_text(text):
    """Preprocess the text with multiple cleaning steps"""
    text = str(text)  # Ensure text is string
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^\w\s.]', '', text)
    text = ' '.join(text.split())
    return text
```

```python
def clean_dataset(df):
    """
    Clean and preprocess the dataset
    """
    # Create copies of the original columns
    df = df.copy()

    # Remove empty or null values
    df = df.dropna(subset=['Content', 'Human Summary'])

    # Remove duplicates
    df = df.drop_duplicates(subset=['Content'])

    # Apply preprocessing to content and summary
    df['Content'] = df['Content'].apply(preprocess_text)
    df['Human Summary'] = df['Human Summary'].apply(preprocess_text)

    # Remove entries with very short content or summaries
    df = df[df['Content'].str.split().str.len() > 10]
    df = df[df['Human Summary'].str.split().str.len() > 5]

    return df
```

```python
class NewsSummarizationDataset(Dataset):
    def __init__(self, data, tokenizer, max_length=256):
        self.tokenizer = tokenizer
        self.data = data
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        input_text = f"summarize: {item['Content']}"

        inputs = self.tokenizer(
            input_text,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        targets = self.tokenizer(
            item['Human Summary'],
            max_length=64,
            padding='max_length',
            truncation=True,
            return_tensors="pt"
        )

        return {
            'input_ids': inputs['input_ids'].squeeze(),
            'attention_mask': inputs['attention_mask'].squeeze(),
            'labels': targets['input_ids'].squeeze()
        }


```

```python
def prepare_data(df):
    """
    Prepare and split the dataset
    """
    # Clean and preprocess the dataset
    df = clean_dataset(df)

    # Take a smaller subset for faster training if needed
    if len(df) > 1000:
        df = df.sample(n=1000, random_state=42)

    # Split data
    train_data, temp_data = train_test_split(df, test_size=0.2, random_state=42)
    val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)

    print(f"Training samples: {len(train_data)}")
    print(f"Validation samples: {len(val_data)}")
    print(f"Test samples: {len(test_data)}")

    return train_data, val_data, test_data
```

```python
def compute_metrics(pred):
    scorer = rouge_scorer.RougeScorer(
        ['rouge1', 'rouge2', 'rougeL'],
        use_stemmer=True
    )

    predictions = tokenizer.batch_decode(pred.predictions, skip_special_tokens=True)
    references = tokenizer.batch_decode(pred.label_ids, skip_special_tokens=True)

    scores = {
        'rouge1': [],
        'rouge2': [],
        'rougeL': []
    }

    for pred, ref in zip(predictions[:100], references[:100]):
        score = scorer.score(ref, pred)
        scores['rouge1'].append(score['rouge1'].fmeasure)
        scores['rouge2'].append(score['rouge2'].fmeasure)
        scores['rougeL'].append(score['rougeL'].fmeasure)

    return {
        'rouge1': np.mean(scores['rouge1']),
        'rouge2': np.mean(scores['rouge2']),
        'rougeL': np.mean(scores['rougeL'])
    }

def get_dataset_statistics(df):
    """
    Get statistics about the dataset
    """
    content_lengths = df['Content'].str.split().str.len()
    summary_lengths = df['Human Summary'].str.split().str.len()

    stats = {
        'Total samples': len(df),
        'Average content length': content_lengths.mean(),
        'Average summary length': summary_lengths.mean(),
        'Max content length': content_lengths.max(),
        'Max summary length': summary_lengths.max()
    }

    return stats
```

```python
import pandas as pd
import numpy as np
import torch
import re
from torch.utils.data import Dataset
from transformers import (
    T5Tokenizer,
    T5ForConditionalGeneration,
    Seq2SeqTrainer,
    Seq2SeqTrainingArguments,
    DataCollatorForSeq2Seq
)
from sklearn.model_selection import train_test_split
from datasets import Dataset as HFDataset
from rouge_score import rouge_scorer
import nltk
nltk.download('punkt', quiet=True)

def preprocess_text(text):
    """Preprocess the text with multiple cleaning steps"""
    text = str(text)
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^\w\s.]', '', text)
    text = ' '.join(text.split())
    return text

def tokenize_data(examples, tokenizer, max_input_length=512, max_target_length=128):
    """Tokenize the data for T5 model"""
    inputs = ["summarize: " + str(doc) for doc in examples["Content"]]
    targets = [str(summary) for summary in examples["Human Summary"]]

    model_inputs = tokenizer(
        inputs,
        max_length=max_input_length,
        padding="max_length",
        truncation=True
    )

    # Tokenize targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            targets,
            max_length=max_target_length,
            padding="max_length",
            truncation=True
        )

    model_inputs["labels"] = labels["input_ids"]

    # Replace padding token id in the labels by -100 to ignore padding in the loss
    model_inputs["labels"] = [
        [-100 if token == tokenizer.pad_token_id else token for token in label]
        for label in model_inputs["labels"]
    ]

    return model_inputs

def prepare_data(df):
    """Prepare and split the dataset"""
    # Ensure columns exist and are properly named
    required_columns = ['Content', 'Human Summary']
    if not all(col in df.columns for col in required_columns):
        raise ValueError(f"DataFrame must contain columns: {required_columns}")

    # Keep only required columns
    df = df[required_columns].copy()

    df['Content'] = df['Content'].apply(preprocess_text)
    df['Human Summary'] = df['Human Summary'].apply(preprocess_text)

    # Remove any rows with empty content or summaries
    df = df.dropna(subset=['Content', 'Human Summary'])

    train_data, temp_data = train_test_split(df, test_size=0.2, random_state=42)
    val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)

    print(f"Training samples: {len(train_data)}")
    print(f"Validation samples: {len(val_data)}")
    print(f"Test samples: {len(test_data)}")

    return train_data, val_data, test_data

def compute_metrics(pred):
    """Compute ROUGE metrics for the summarization task"""
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    # Get predictions and labels
    predictions = pred.predictions
    labels = pred.label_ids

    # Decode predictions and labels
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)

    # Replace -100 in labels with pad_token_id
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Remove extra whitespace
    decoded_preds = [pred.strip() for pred in decoded_preds]
    decoded_labels = [label.strip() for label in decoded_labels]

    # Compute ROUGE scores
    scores = {
        'rouge1': [],
        'rouge2': [],
        'rougeL': []
    }

    for pred, label in zip(decoded_preds, decoded_labels):
        score = scorer.score(label, pred)
        scores['rouge1'].append(score['rouge1'].fmeasure)
        scores['rouge2'].append(score['rouge2'].fmeasure)
        scores['rougeL'].append(score['rougeL'].fmeasure)

    return {
        'rouge1': np.mean(scores['rouge1']),
        'rouge2': np.mean(scores['rouge2']),
        'rougeL': np.mean(scores['rougeL'])
    }

def generate_summary(article_text, model, tokenizer):
    """Generate a summary for the given article text"""
    inputs = tokenizer(
        "summarize: " + preprocess_text(article_text),
        return_tensors="pt",
        max_length=512,
        padding=True,
        truncation=True
    )

    summary_ids = model.generate(
        inputs["input_ids"],
        num_beams=4,
        min_length=30,
        max_length=128,
        early_stopping=True
    )
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)

if __name__ == "__main__":
    # Initialize tokenizer and model
    model_name = 't5-small'
    tokenizer = T5Tokenizer.from_pretrained(model_name)
    model = T5ForConditionalGeneration.from_pretrained(model_name)

    # Load data
    df = pd.read_excel("/content/Synthetic News Dataset & Model Details.xlsx", sheet_name=0)
    df.columns = df.iloc[0]
    df = df.drop([0]).reset_index(drop=True)
    df = df.rename_axis(None, axis=1)

    # Prepare datasets
    train_data, val_data, test_data = prepare_data(df)

    # Convert to HuggingFace datasets
    train_dataset = HFDataset.from_pandas(train_data)
    val_dataset = HFDataset.from_pandas(val_data)
    test_dataset = HFDataset.from_pandas(test_data)

    # Apply tokenization
    train_dataset = train_dataset.map(
        lambda x: tokenize_data(x, tokenizer),
        batched=True,
        remove_columns=train_dataset.column_names
    )

    val_dataset = val_dataset.map(
        lambda x: tokenize_data(x, tokenizer),
        batched=True,
        remove_columns=val_dataset.column_names
    )

    test_dataset = test_dataset.map(
        lambda x: tokenize_data(x, tokenizer),
        batched=True,
        remove_columns=test_dataset.column_names
    )

    # Create data collator
    data_collator = DataCollatorForSeq2Seq(
        tokenizer=tokenizer,
        model=model
    )

    # Training arguments using Seq2SeqTrainingArguments
    training_args = Seq2SeqTrainingArguments(
        output_dir="./t5-news-summarizer",
        num_train_epochs=3,
        per_device_train_batch_size=4,
        per_device_eval_batch_size=4,
        warmup_steps=50,
        weight_decay=0.01,
        logging_dir='./logs',
        logging_steps=10,
        evaluation_strategy="steps",
        eval_steps=20,
        save_steps=20,
        save_total_limit=1,
        fp16=torch.cuda.is_available(),
        gradient_accumulation_steps=4,
        predict_with_generate=True,
        generation_max_length=128,
        generation_num_beams=4,
        report_to="none"
    )

    # Initialize trainer using Seq2SeqTrainer
    trainer = Seq2SeqTrainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        data_collator=data_collator,
        compute_metrics=compute_metrics,
        tokenizer=tokenizer  # Added tokenizer here for generation
    )

    # Train the model
    trainer.train()

    # Evaluate
    test_results = trainer.evaluate(test_dataset)
    print(f"\nTest Results: {test_results}")

    # Save the model
    model.save_pretrained("./t5-news-summarizer-final")
    tokenizer.save_pretrained("./t5-news-summarizer-final")

    # Generate sample summary
    if len(test_data) > 0:
        sample_article = test_data.iloc[0]['Content']
        print("\nSample Summary Generation:")
        print("Original article:", sample_article[:200], "...")
        print("Generated summary:", generate_summary(sample_article, model, tokenizer))
        print("Original summary:", test_data.iloc[0]['Human Summary'])
```

```python
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration
import re
from typing import Optional
import uvicorn

app = FastAPI(
    title="News Summarization API",
    description="API for generating summaries from news articles using a fine-tuned T5 model",
    version="1.0.0"
)

class ArticleInput(BaseModel):
    text: str
    min_length: Optional[int] = 30
    max_length: Optional[int] = 128

class SummaryOutput(BaseModel):
    summary: str
    original_length: int
    summary_length: int

def preprocess_text(text: str) -> str:
    """Preprocess the text with multiple cleaning steps"""
    text = str(text)
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[^\w\s.]', '', text)
    text = ' '.join(text.split())
    return text

class SummarizationModel:
    def __init__(self, model_path: str):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.tokenizer = T5Tokenizer.from_pretrained(model_path)
        self.model = T5ForConditionalGeneration.from_pretrained(model_path).to(self.device)
        self.model.eval()

    def generate_summary(self, text: str, min_length: int = 30, max_length: int = 128) -> str:
        inputs = self.tokenizer(
            "summarize: " + preprocess_text(text),
            return_tensors="pt",
            max_length=512,
            padding=True,
            truncation=True
        )

        inputs = {k: v.to(self.device) for k, v in inputs.items()}

        with torch.no_grad():
            summary_ids = self.model.generate(
                inputs["input_ids"],
                num_beams=4,
                min_length=min_length,
                max_length=max_length,
                early_stopping=True
            )

        return self.tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Initialize model
MODEL_PATH = "/content/t5-news-summarizer-final"  # Update this to your model path
summarizer = SummarizationModel(MODEL_PATH)

@app.get("/")
def read_root():
    return {
        "message": "Welcome to the News Summarization API",
        "version": "1.0.0",
        "endpoints": {
            "/summarize": "POST request to generate summary",
            "/health": "GET request to check API health"
        }
    }

@app.get("/health")
def health_check():
    return {"status": "healthy", "model_loaded": True}

@app.post("/summarize", response_model=SummaryOutput)
def generate_summary(article: ArticleInput):
    try:
        # Validate input
        if not article.text.strip():
            raise HTTPException(status_code=400, detail="Empty article text provided")

        if len(article.text.split()) < 10:
            raise HTTPException(status_code=400, detail="Article text too short")

        # Generate summary
        summary = summarizer.generate_summary(
            article.text,
            min_length=article.min_length,
            max_length=article.max_length
        )

        return SummaryOutput(
            summary=summary,
            original_length=len(article.text.split()),
            summary_length=len(summary.split())
        )

    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

if __name__ == "__main__":
    uvicorn.run("app:app", host="0.0.0.0", port=8000, reload=True)
```
